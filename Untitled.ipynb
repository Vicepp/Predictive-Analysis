```python
miss_cols=["PoolQC","MiscFeature","Alley","Fence","FireplaceQu","GarageType","GarageFinish","GarageQual","GarageCond",
                  "BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1","BsmtFinType2","MasVnrType","MSSubClass"]
for i in miss_cols:
    df[i].fillna("none", inplace=True)
```

```python
df["Functional"] = df["Functional"].fillna("typ")
```

```python
miss_cols2= ["MSZoning", "BsmtFullBath", "BsmtHalfBath", "Utilities","MSZoning",
                      "Electrical", "KitchenQual", "SaleType","Exterior1st", "Exterior2nd","MasVnrArea"] 
for i in miss_cols2:
    df[i].fillna(df[i].mode()[0], inplace=True)
```

```python

df["LotFrontage"]= df.groupby("Neighborhood")["LotFrontage"].transform(lambda x : x.fillna(x.median()))
```

```python
df.isna().sum().nlargest(5)
```

```python

nm  = ["MSSubClass", "MoSold", "YrSold"]
for col in nm:
    df[col].astype(str)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
ibe= LabelEncoder()
encodecolumns = ("FireplaceQu","BsmtQual","BsmtCond","ExterQual","ExterCond","HeatingQC","GarageQual",
                "GarageCond","PoolQC","KitchenQual","BsmtFinType1","BsmtFinType2","Functional","Fence",
                "BsmtExposure","GarageFinish","LandSlope","LotShape","PavedDrive","Street","Alley",
                "CentralAir","MSSubClass","OverallCond","YrSold","MoSold")
for i in encodecolumns :
    ibe.fit(list(df[i].values))
    df[i] = ibe.transform(list(df[i].values))
```

```python
from scipy.stats import skew
```

```python
num = df.dtypes[df.dtypes !="object"].index
skewed_var= df[num].apply(lambda x:skew(x.dropna())).sort_values(ascending= False)
skw=pd.DataFrame({"Skewed Features": skewed_var})
skw
```

```python
from scipy.special import boxcox1p
```

```python
skw= skw[abs(skw)>0.75]
skw_var2= skw.index
for i in skw_var2:
    df[i]= boxcox1p(df[i], 0.15)
    df[i] +=1
```
